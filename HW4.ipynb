The OLS had an R-squared of .833, which means that it was decently fit. R-squared is a measure that tracks how close the data fits the line made by the model, with 1 being a perfect score.

The co-efficients show the weights of the different independent variables, and how much they influence the dependent variable. Negative ones have a negative correlation, and positive ones have a positive correlation. The coefficients fit in the linear regression formula as the B1, B2, B3, and so on.

We can tell which independent variables are more significant by looking at the P>|t| column. Columns with a value lower than .05 will be significant, as this column measures how likely the outcome is a result of random chance. Lower values mean that its not because of chance, and are an actual outcome. In this regression, we see that likes, shares, users, and consumers are lower than .05, and thus significant.

```python
#kfold with alpha 0
errlist = []
Rerrlist = []


kf = KFold(n_splits=5, shuffle=True)
for train_index, test_index in kf.split(copyX):
    X_train, X_test = copyX.iloc[train_index], copyX.iloc[test_index]
    y_train, y_test = copyY.iloc[train_index], copyY.iloc[test_index]
    model = sm.OLS(y_train.values, X_train.values)
    results = model.fit()
    predict = results.predict(X_test)
    predict = predict.clip(lower=0)

    mse = mean_squared_error(y_test.values, predict.values)
    errlist.append(mse)

    resultsR = model.fit_regularized(method='elastic_net', L1_wt=0.0, alpha=0)
    predictR = resultsR.predict(X_test)
    predictR = predictR.clip(lower=0)

    mseR = mean_squared_error(y_test.values, predictR.values)

    Rerrlist.append(mseR)
    
print("All MSE for regular fit: ", errlist)
print("Average MSE for regular fit: ", np.mean(errlist))
print("All MSE for fit_regularized: ", Rerrlist)
print("Average MSE for fit_regularized: ", np.mean(Rerrlist))
```

```python
errlist = []
Rerrlist = []


kf = KFold(n_splits=5, shuffle=True)
for train_index, test_index in kf.split(labeledX):
    X_train, X_test = labeledX.iloc[train_index], labeledX.iloc[test_index]
    y_train, y_test = labeledY.iloc[train_index], labeledY.iloc[test_index]
    model = sm.OLS(y_train.values, X_train.values)
    results = model.fit()
    predict = results.predict(X_test)
    predict = predict.clip(lower=0)
    mse = mean_squared_error(y_test.values, predict.values)
    errlist.append(mse)
    
    whole = X_train.copy()
    whole['comment'] = y_train
    
    
    modelR = smf.ols(formula="comment ~ np.power(like, 2) + np.power(share, 2) + like + share", data=whole)
    resultsR = modelR.fit_regularized(method='elastic_net', L1_wt=0.0, alpha=10)
    predictR = resultsR.predict(X_test)
    predictR = predictR.clip(lower=0)
    mseR = mean_squared_error(y_test.values, predictR.values)
    
    Rerrlist.append(mseR)
    
print("All MSE for regular fit: ", errlist)
print("Average MSE for regular fit: ", np.mean(errlist))
print("All MSE for fit_regularized: ", Rerrlist)
print("Average MSE for fit_regularized: ", np.mean(Rerrlist))
```
